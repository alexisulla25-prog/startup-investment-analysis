```python
import pandas as pd

# Esta es la URL correcta para descargar SOLO los datos
url = 'https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/Startups%20in%202021%20end.csv'

# Cambiamos el sep=";" por sep="," porque es un CSV estÃ¡ndar
startups = pd.read_csv(url,
                       encoding='ISO-8859-1',
                       on_bad_lines="skip",
                       sep=",")

# Verificamos que cargÃ³ bien
startups.head()
```

```python
startups.dtypes
```

```python
#cambiamos por numero flotante para poder sumar
startups['Valuation ($B)'] = startups['Valuation ($B)'].str.replace('$', '').str.replace('B', '', regex=False)
startups['Valuation ($B)'] = pd.to_numeric(startups['Valuation ($B)'], errors='coerce')
```

```python
startups.head() #obervamos que se haya creado correctamente
```

```python
#cambiamos el nombre de la columna unnamed para no generar confusion
startups.rename(columns={'Unnamed: 0': 'company_id'}, inplace=True)
```

```python
#Indicadores y metricas clave del data set

#Empresas de IA en China
mask=startups['Industry']=='Artificial intelligence'
mask2=startups['Country']=='China'
startups[mask & mask2].reset_index(drop=True)
```

```python
#Empresa con mayor valuacion
max_valuation = startups['Valuation ($B)'].max()
company_with_max_valuation = startups[startups['Valuation ($B)'] == max_valuation]
display(company_with_max_valuation)
```

```python
# Numero total de paises unicos
total_countries = startups['Country'].nunique()
print(f"Numero total de paises: {total_countries}")
```

```python
# Empresas por pais
companies_per_country = startups['Country'].value_counts()
display(companies_per_country)
```

```python
# Numero total de industrias unicas
total_industries = startups['Industry'].nunique()
print(f"Numero total de industrias: {total_industries}")
```

```python
# Empresas por industria
companies_per_industry = startups['Industry'].value_counts()
display(companies_per_industry)
```

```python
# Guardar el DataFrame 'startups' en un archivo CSV
startups.to_csv('startups.csv', index=False)
print("DataFrame 'startups' guardado como 'startups.csv'")
```
